Iris data を読み込む⇒データフレームを作る

```python
# 1. データの読み込み
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
feature_names = iris.feature_names

# データフレームをつくる
df = pd.DataFrame(data=X, columns=feature_names)
df["target"] = y
display(df.head())

print(X.shape)
print(y.shape)
print(target_names)
df["target"].value_counts()
```

```python
import seaborn as sns

sns.pairplot(df, hue="target", markers=["o", "^", "D"], palette=["red", "green", "blue"])
plt.show()
```

```python
# ヒストグラム "sepal length (cm)"

color = {0: 'red', 1: 'green', 2: 'blue'}
label = ['Setosa',  'Versicolor', 'Virginica']
column = ["sepal length (cm)",	"sepal width (cm)",
          "petal length (cm)",	"petal width (cm)"]


plt.figure(figsize=[8, 4])

for target in [0, 1, 2]:
  subset = df[df['target'] == target]
  plt.hist(subset[column[0]], alpha=0.5, bins=9, edgecolor='black',
         color=color[target], label=label[target])

plt.xlabel(column[0], fontsize=14)
plt.xticks(fontsize=14)
plt.xlim(1, 10)
plt.ylabel('Frequency', fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(0, 20)
plt.grid()
plt.legend()
plt.show()
```
