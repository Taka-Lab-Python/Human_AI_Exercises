Kerasのモジュールのインポート

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

kerasのモジュールの中にあるMNISTデータセットをインポートする (60000, 28, 28)

```python
# MNIST データセットの表示
(X, y), _ = tf.keras.datasets.mnist.load_data()
# 画像データ
print(X.shape)

# ラベル　教師データ
print(y.shape)

# データ可視化
n, col = 20, 5
line = int(n/col)
plt.figure(figsize=(8, 8))
for i in range(n):
  plt.subplot(line, col, i+1)
  plt.axis("off")
  plt.imshow(X[i], cmap=plt.cm.gray_r,interpolation="nearest")
  plt.title(y[i])
plt.show()
```
