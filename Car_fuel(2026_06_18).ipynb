```python
# 線形単回帰でR2決定係数を求める

from sklearn.linear_model import LinearRegression

X = df[["燃料消費 Liter"]]
y = df["走行距離 km"]

model = LinearRegression()
model.fit(X, y)

r2 = model.score(X, y)
a = model.coef_[0]
b = model.intercept_

print(f"a={a:.4f}, b={b:.4f}, R2={r2:.4f}")
```
