```python
import numpy as np
num = 500
lim = 2
x_data = [np.random.uniform(-lim, lim) for _ in range(num)]
y_data = [np.random.uniform(-lim, lim*2) for _ in range(num)]
P_data = np.array([x_data, y_data]).T

f = lambda x : x**2

t_data = (P_data[:,1] > f(P_data[:,0])).astype(np.int32)

plt.figure(figsize=[4, 3])
x = np.linspace(-lim, lim, num)
plt.plot(x, f(x), c="black")

for i in range(num):
  if t_data[i] == 0:
    plt.scatter(x_data[i], y_data[i], s = 20, marker=".", c="blue")

plt.grid()
plt.show()

#t_data = (P_data[:,1] > f(P_data[:,0])).astype(np.int32)
t_data = (P_data[:,1] > f(P_data[:,0]))
```
