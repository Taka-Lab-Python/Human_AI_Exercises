Classを使わないパーセプトロンのプログラム　ORゲート

```python
# パーセプトロン
import numpy as np
import time

# --- 1. データと「初期設定（パラメータ）」 ---
# ORゲートのデータ（入力 X と 正解 y）
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y_true = np.array([0, 1, 1, 1])  # 正解ラベル

# 初期状態（最初は重みもバイアスも全部 0 からスタート）
w1 = 0.0
w2 = 0.0
b = 0.0
lr = 0.1  # 学習率

print("【学習スタート前の状態】")
print(f"初期値 -> 重み1: {w1}, 重み2: {w2}, バイアス: {b}\n")
print("==========================================")

# --- 2. 学習スタート（4つのデータを4回繰り返し見せる） ---
epochs = 4

for epoch in range(epochs):
    print(f"--- \n データの確認 {epoch + 1}エポック  ---")

    for i in range(4):
        # データを1個ずつ取り出す
        x1 = X[i][0]
        x2 = X[i][1]
        target = y_true[i]

        # 【計算ステップ】掛け算して足して、バイアスを足す
        total = (x1 * w1) + (x2 * w2) + b

        # 【活性化関数（フィルター）】0以上なら1、マイナスなら0
        if total >= 0:
            prediction = 1
        else:
            prediction = 0

        # 【答え合わせ】正解と予測のズレを計算
        error = target - prediction

        # 【つまみの調整】間違えていたら、ここで重みとバイアスを書き換える！
        if error != 0:
            w1 = w1 + (lr * error * x1)
            w2 = w2 + (lr * error * x2)
            b = b + (lr * error)
            result_str = "❌ 不正解！"
        else:
            result_str = "⭕ 正解！"

        # 1ステップずつの動きを画面に出力
        print(f"入力:[{x1}, {x2}] | 正解:{target} | 予測:{prediction} -> {result_str} (現在の w1:{w1:.1f}, w2:{w2:.1f}, b:{b:.1f})")
        time.sleep(0.1) # 時間キープ

print("\n==========================================")
print("【学習完了！】収束値")
print(f"収束値 -> 重み1: {w1:.1f}, 重み2: {w2:.1f}, バイアス: {b:.1f}")
```
