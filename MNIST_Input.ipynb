```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier

# ==========================================
#  MNIST データを読み込む
# ==========================================
digits = load_digits()
X = digits.data         # 1次元画像データ
images = digits.images  # 2次元画像データ
y = digits.target       # 教師数字 (0～9)

print(f"データ全体の数: {len(X)}件")

# ==========================================
# 　トレーニングデータとテストデータに分ける
# ==========================================

split_index = 1500  #分割ポイント

X_train = X[:split_index]  # 0番目から1499番目まで（練習問題）
y_train = y[:split_index]  # その正解

X_test = X[split_index:]  # 1500番目から最後まで（本番テスト）
y_test = y[split_index:]  # その正解

print(f"トレーニング（練習用）の数: {len(X_train)}件")
print(f"テストの数: {len(X_test)}件")

# ==========================================
#  MNISTデータ
# X_train, y_train, X_test, y_test
# ==========================================
print(y_test)
```
