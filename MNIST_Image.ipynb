3. MNIST 数値データを可視化する

```python
# MNISTを可視化する
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

images = digits.images

plt.figure(figsize=(12, 5))
plt.suptitle("MNIST Dataset", fontsize=14, fontweight='bold')

# 3. ループを使って横に画像を並べる
for i in range(10):
    plt.subplot(2, 5, i + 1)

    # 画像を表示（cmap='gray_r' で白黒反転）
    plt.imshow(images[i])

    # 各画像の上に教師データ（ラベル）を表示
    plt.title(f"(y): {y[i]}", fontsize=14)

    # グラフの目盛り（0〜7の数字）を非表示にして見やすくする
    plt.axis('off')

# 画面に表示
plt.tight_layout() # 上下で重ならないように隙間をあける
plt.show()
```
