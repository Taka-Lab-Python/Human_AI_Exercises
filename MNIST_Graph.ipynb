```python
# 8. LossとAccuracyのグラフ描画
from matplotlib.ticker import MaxNLocator

plt.figure(figsize=(10, 8))

# Loss（損失）の推移
ax1 = plt.subplot(2, 1, 2)
plt.plot(history.history['loss'], label='Train Loss', color='blue', linewidth=3)
plt.plot(history.history['val_loss'], label='Val Loss', color='red', linestyle='--', linewidth=3)
plt.title('Model Loss(損失)', fontsize=16)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend()
plt.grid(True)
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))

# Accuracy（精度）の推移
plt.subplot(2, 1, 1)
plt.plot(history.history['accuracy'], label='Train Acc', color='blue', linewidth=3)
plt.plot(history.history['val_accuracy'], label='Val Acc', color='red', linestyle='--', linewidth=3)
plt.title('Model Accuracy(精度)', fontsize=16)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend()
plt.grid(True)
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.tight_layout()
plt.show()
```
