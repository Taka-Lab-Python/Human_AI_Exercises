```python
# ==========================================
# 5. 予測と「正解率」の計算
# ==========================================

# ① 練習問題に対する予測（学習できているかの確認）
train_pred = model.predict(X_train)

# ② テストに対する予測（実力の確認）
test_pred = model.predict(X_test)

#  ① 練習問題、正解数をカウントする（if文で判定させる）
train_correct = 0
for i in range(len(train_pred)):
    if train_pred[i] == y_train[i]:
        train_correct += 1

#  ② テスト問題、正解数をカウントする（if文で判定させる）
test_correct = 0
for i in range(len(test_pred)):
    if test_pred[i] == y_test[i]:
        test_correct += 1

# 正解率（％）を計算する
train_accuracy = (train_correct / len(X_train)) * 100
test_accuracy = (test_correct / len(X_test)) * 100

print("練習問題の正解率（％）{:.1f}:". format(train_accuracy))
print("テスト問題の正解率（％）{:.1f}:". format(test_accuracy))
```
