scikit learnライブラリのインポート

```python
# Scikit Learnライブラリのインポート
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

【ステップ 1: データの準備（インポート）】

```python
# ==========================================
# 【ステップ 1: データの準備】
# ==========================================
iris = load_iris()
X = iris.data  # 特徴量（がくの長さ/幅、花弁の長さ/幅の4つ）
y = iris.target  # 正解ラベル（0: Setosa, 1: Versicolor, 2: Virginica
class_names = iris.target_names

# 出力
# 1. データ数と特徴量の数（shapeを使用）
num_samples, num_features = X.shape

print(f"データ数（サンプルの総数）: {num_samples} 件")
print(f"特徴量の数: {num_features} 個")

# 2. それぞれの正解ラベルの数（np.uniqueを使用）
labels, counts = np.unique(y, return_counts=True)

print("\nそれぞれの正解ラベルの数:")
for name, count in zip(iris.target_names, counts):
    print(f" - {name} : {count} 件")
```

【ステップ 2: データの分割（訓練用とテスト用）】

```python
# ==========================================
# 【ステップ 2: データの分割】
# ==========================================
# データの70%を訓練用、30%をテスト用に分割します
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

【ステップ 3: モデルの定義（アルゴリズムの選択）】

 ▼ 決定木（DecisionTreeClassifier）を選択し、木の深さなどの設定を行う

```python
# ==========================================
# 【ステップ 3: モデルの定義】max_depth=木の深さ
# ==========================================
model = DecisionTreeClassifier(max_depth=2, random_state=42)
```

【ステップ 4: 学習（fit）】

▼ 訓練データをモデルに入れる、機械に自動で分類ルールを見つけさせる

```python
# ==========================================
# 【ステップ 4: 学習（fit）】
# ==========================================
# 訓練データを与えて、機械にルールを自動学習させます
model.fit(X_train, y_train)
print("学習が完了しました。機械が自動でルールを構築しました。")
```

ステップ 5: 予測と評価（predict & 精度計算）

```python
# ステップ 5: 予測と評価（predict & 精度計算）
# ==========================================
# 【ステップ 5: 予測と評価】
# ==========================================
# テストデータを使って予測
y_pred = model.predict(X_test)

# 正解率（精度）の計算
accuracy = accuracy_score(y_test, y_pred)
print(f"テストデータでの正解率: {accuracy:.2f} ({accuracy * 100:.1f}%)")
```

ステップ６：可視化（plot_tree）

```python
# ==========================================
# 【可視化: 自動生成されたルールの表示】
# ==========================================
plt.figure(figsize=(10, 7))
plot_tree(
    model,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,  # 分類ごとに色を塗る
    rounded=True,  # ノードの角を丸くする
)
plt.show()
```

【ステップ 7: 混同行列の可視化】

```python
# ==========================================
# 【ステップ 7: 混同行列と決定木の可視化】
# ==========================================
# 全体のフォントサイズを大きく設定

from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
plt.rcParams.update({'font.size': 14})

# --- 1つ目：混同行列 (Confusion Matrix) ---
fig_cm, ax_cm = plt.subplots(figsize=(7, 5))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot(cmap=plt.cm.Blues, ax=ax_cm)

ax_cm.set_title('混同行列(こんどうぎょうれつ)', fontsize=18)

# 四角の中の数字をさらに大きく読みやすくする
for text_obj in disp.text_.ravel():
    text_obj.set_fontsize(24)
```

【ステップ 8: 分類結果の可視化】

```python
# ステップ 8: 分類結果の可視化

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

# 1. データの読み込み
iris = load_iris()
X = iris.data
y = iris.target

# 2. 決定木モデルの学習（max_depth=2）
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X, y)

# 3. 予測
y_pred = clf.predict(X)
# 4. 不正解データのインデックスを取得
incorrect_indices = ( y != y_pred)

# 5. 可視化（特徴量：花弁の長さと幅を利用）
plt.figure(figsize=(10, 6))

# 正解データをプロット（薄い色で）
plt.scatter(X[~incorrect_indices, 2], X[~incorrect_indices, 3],
            c=y[~incorrect_indices], cmap='viridis', s=50, label='正解だったよ', alpha=0.5)

# 不正解データを赤色でプロット
plt.scatter(X[incorrect_indices, 2], X[incorrect_indices, 3],
            color='red', marker='x', s=150, linewidths=3, label='不正解でした(残念)')

plt.title("Iris データの分類(Decision Tree, max_depth=2)")
plt.xlabel("Petal Length (cm)", fontsize=16)
plt.ylabel("Petal Width (cm)", fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```
